```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
```

```python
flights = pd.read_csv(
    r"C:\Users\ACER\Downloads\flights.csv\flights.csv",
    low_memory=False
)

airlines = pd.read_csv(r"C:\Users\ACER\Downloads\airlines.csv")

airports = pd.read_csv(r"C:\Users\ACER\Downloads\airports.csv")
```

```python
flights_clean = flights.copy()
airlines_clean = airlines.copy()
airports_clean = airports.copy()
```

```python
missing = flights_clean.isnull().sum()

missing_percentage = (missing / len(flights_clean)) * 100

missing_table = pd.DataFrame({
    "Missing Values": missing,
    "Missing Percentage": missing_percentage
})

missing_table.sort_values(by="Missing Percentage", ascending=False)
```

```python
flights_clean = flights_clean.drop_duplicates()
```

```python
flights_clean.dtypes
```

```python
flights_clean.info()
```

```python
flights_clean.columns = flights_clean.columns.str.lower()
airlines_clean.columns = airlines_clean.columns.str.lower()
airports_clean.columns = airports_clean.columns.str.lower()
```

```python
for col in flights_clean.select_dtypes(include=['object', 'string']).columns:
    flights_clean[col] = flights_clean[col].str.strip()
```

```python
for col in airlines_clean.select_dtypes(include=['object', 'string']).columns:
    airlines_clean[col] = airlines_clean[col].str.strip()

for col in airports_clean.select_dtypes(include=['object', 'string']).columns:
    airports_clean[col] = airports_clean[col].str.strip()
```

```python
flights_clean["airline"] = flights_clean["airline"].str.upper()

flights_clean["origin_airport"] = flights_clean["origin_airport"].str.upper()

flights_clean["destination_airport"] = flights_clean["destination_airport"].str.upper()
```
